#Import Library

```python
import pandas as pd
import numpy as np
import csv
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

#Load Dataset

```python
df = pd.read_csv('/content/SMILES_Big_Data_Set.csv')
```

#Preprocessing

Checking Missing Values pada dataset.

```python
missing_values = df.isnull().sum()
print(missing_values)
```

Checking baris data Duplikat.

```python
df.duplicated().sum()
```

Feature Scaling untuk fitur Numerik.

```python
df_numerical = df.select_dtypes(include=['float64', 'int64']).columns

standard_scaler = StandardScaler()
df[df_numerical] = standard_scaler.fit_transform(df[df_numerical])

df.head()
```

Menghapus fitur Mol, karena tidak dibutuhkan.

```python
df = df.drop(columns='mol')
df.head()
```

Feature Encoding untuk fitur Kategorikal, yaitu SMILES.

```python
df_categorical = df.select_dtypes(include=['object']).columns

encoder = LabelEncoder()

for col in df_categorical:
    df[col] = encoder.fit_transform(df[col])

df.head()
```

Menghapus baris data yang memiliki nilai kosong.

```python
df.dropna(inplace=True)

print(df.isnull().sum())
```

Handle Outlier untuk kolom numerik, kecuali SMILES.

```python
for col in df_numerical:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

df.head()
```

Save Dataset Setelah Melewati Preprocessing.

```python
df.to_csv('preprocessed_SMILES_Big_Data_Set.csv', index=False)
print("Dataset has been saved to 'preprocessed_SMILES_Big_Data_Set.csv'")
```
